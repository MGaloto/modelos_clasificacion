```python
# Defino el nombre del archivo de entrada
input_file = "winemag-data_first150k.csv"
df_pandas = pd.read_csv(input_folder + input_file, index_col=0)
df_pandas['id'] = df_pandas.index
df_pandas
```

#### Correlación
En probabilidad y estadística, la **correlación** indica la fuerza y la dirección de una relación lineal y proporcionalidad entre dos variables estadísticas. 
Se considera que dos variables cuantitativas están correlacionadas cuando los valores de una de ellas varían sistemáticamente con respecto a los valores homónimos de la otra: si tenemos dos variables (A y B) existe correlación entre ellas si al disminuir los valores de A lo hacen también los de B y viceversa.

### Machine Learning - Gradient Boosted Trees (GBT)
Gradient boosting es una técnica de machine learning orientada tanto a problemas de regresión como de clasificación. Produce un modelo de predicción del tipo "ensable", típicamente utilizando un árbol de decisión.

```python
### Es necesario convertir columnas categoricas a numéricas
df_pandas_pred = df_pandas 
df_pandas_pred['country_num'] = df_pandas_pred['country'].astype('category').cat.codes
df_pandas_pred['designation_num'] = df_pandas_pred['designation'].astype('category').cat.codes
df_pandas_pred['province_num'] = df_pandas_pred['province'].astype('category').cat.codes
df_pandas_pred['region_1_num'] = df_pandas_pred['region_1'].astype('category').cat.codes
df_pandas_pred['variety_num'] = df_pandas_pred['variety'].astype('category').cat.codes
df_pandas_pred['winery_num'] = df_pandas_pred['winery'].astype('category').cat.codes
```

```python
df_pandas_pred
```

```python
df_pandas_pred.describe()
```

```python
df_pandas_pred = df_pandas_pred[["country_num", "designation_num", "province_num", "region_1_num", "variety_num", "winery_num", "price", "points"]]
df = spark.createDataFrame(df_pandas_pred)
train, test = df.randomSplit([0.7, 0.3])
print ("Tenemos %d observaciones para el entrenamiento y %d para prueba/testing." %(train.count(), test.count()))
```

```python
from pyspark.ml.regression import GBTRegressor

# Definimos el modelo -> Gradient Boosted Trees (GBT).
gbt = GBTRegressor(featuresCol='features',
                   labelCol='points',
                   predictionCol="Prediction_points",
                   seed=43)
```

```python
from pyspark.ml.feature import VectorAssembler

featuresCols = df.columns
featuresCols.remove('points')

# Creamos el vector assembler para transformar nuestro dataset
vectorAssembler = VectorAssembler(inputCols=featuresCols, outputCol='features')

# Aplicamos el vector a nuestros datos de entrenamiento y testing
train_df_vec = vectorAssembler.transform(train)
test_df_vec = vectorAssembler.transform(test)
```

```python
train_df_vec.show()
```

```python
# Entrenamiento del modelo. El modelo se guardará en gbt_model
gbt_model = gbt.fit(train_df_vec)
```

```python
# Usamos el modelo generado en gbt_model para aplicarlo en el set de testing
gbt_predictions = gbt_model.transform(test_df_vec)
```

```python
# En la columna Prediction_points vemos el resultado de aplicar el modelo, y lo podemos comparar con la columna "points"
gbt_predictions.select("points", "Prediction_points", *featuresCols).show()
```

#### Evaluación de los resultados
*RMSE es un valor estadístico para evaluar el accuracy de nuestro modelo. Se calcula a través de la desviación estándard de la diferencia entre los valores de la predicción y las observaciones.*

```python
from pyspark.ml.evaluation import RegressionEvaluator

# Select columns to compute test error
evaluator = RegressionEvaluator(labelCol='points', 
                                predictionCol='Prediction_points')
rmse = evaluator.evaluate(gbt_predictions, {evaluator.metricName: 'rmse'})

# Print Model Metrics
print(' RMSE: ' + str(rmse))
#RMSE: 2.445891024290397
```

### Analizamos la importancia de cada variable en el modelo

```python
# Convertimos el vector de variables de importancia en un dataframe Pandas
gbt_fi_df = pd.DataFrame(gbt_model.featureImportances.toArray(),
                         columns=['importance'])
gbt_fi_df['feature'] = pd.Series(featuresCols)
# Ordenamos variables por orden de importancia
gbt_fi_df.sort_values(by=['importance'], ascending=False, inplace=True)
```
